```
# Membership Functions - PHI X THETA - Phase 1

def triangle_mem(data,pack):
    (a, m, b,down_co, up_co, stable) = pack
    if stable is not None:
        if stable[0] <= data and data <= stable[1]:
            return stable[2]
    if down_co is not None and data < down_co:
        return 0
    if up_co is not None and up_co < data:
        return 0
    
    if data < a:
        return 0
    if a <= data and data < m:
        return ((data - a )/(m - a))
    if m <= data and data < b:
        return ((b - data )/(b - m))
    if b <= data:
        return 0
    
    

phi = {
    "S3":(-115,-65,-15,-90,None,None),
    "S2":(-45,0,45,None,None,None),
    "S1":(-45,0,45,None,None,None),
    "CE":(15,52.5,90,None,None,None),
    "B1":(80,90,100,None,None,None),
    "B2":(90,127.5,165,None,None,None),
    "B3":(195,245,295,None,270,None)
}

x = {
    "S2":(-4,1.5,7,1.5,None,(0,1.5,1)),
    "S1":(4,7,10,None,None,None),
    "CE":(9,10,11,None,None,None),
    "B1":(10,13,16,None,None,None),
    "B2":(13,18.5,24,None,18.5,(18.5,20,1))
}

theta ={
    "S3":(-60,-40,-20,-40,None,None),
    "S2":(-33,-20,-7,None,None,None),
    "S1":(-14,-7,0,None,None,None),
    "CE":(-4,0,-4,None,None,None),
    "B1":(0,7,14,None,None,None),
    "B2":(7,20,33,None,None,None),
    "B3":(20,40,60,None,40,None)
}
```

For the step 2 and 3, we need to extract fuzzy rules from the datapoints and then eliminate incompatible rules with reference to their associated degrees. 

```
# Rule Extraction - Degree Assignment Phase 2 and 3

rules = []
for dataset in datasets:
    for data in dataset:
        for memx_key, memx_value in x.items():
            for memp_key, memp_value in phi.items():
                for memt_key, memt_value in theta.items():
                    m_x = triangle_mem(data[0],memx_value)
                    m_p = triangle_mem(data[1],memp_value)
                    m_t = triangle_mem(data[2],memt_value)
                    deg = m_x * m_p * m_t
                    if deg > 0 :
                        rules.append(((memx_key,memp_key),memt_key,deg))

# Eliminate incompatible rules, or repetitive rules 
while True:
    len_prev = len(rules)
    
    for rule1 in rules:
        for rule2 in rules:
            if rule1[0] == rule2[0] and rule1[1] != rule2[1]:
                if rule1[2] <= rule2[2]:
                    try:
                        rules.remove(rule1)
                    except:
                        pass
                else:
                    try:
                        rules.remove(rule2)
                    except:
                        pass
            if rule1[0] == rule2[0] and rule1[1] == rule2[1] and rule1[2] != rule2[2]:
                if rule1[2] <= rule2[2]:
                    try:
                        rules.remove(rule1)
                    except:
                        pass
                else:
                    try:
                        rules.remove(rule2)
                    except:
                        pass
    if len_prev == len(rules):
        break
```

As the last step, deffuzification consists of extracting the theta value from rules and a datapoint like (x,\phi). The function below is set to do so. 

```
# Defuzzification - Phase 4

def defuzz(x_test, phi_test):
    numi = 0
    denumi = 0

    for i in range(len(rules)):
        I_1 = rules[i][0][0]
        I_2 = rules[i][0][1]
        O = rules[i][1]
        # print(I_1, I_2, O)
        m_I_1 = triangle_mem(x_test, x[I_1])
        m_I_2 = triangle_mem(phi_test, phi[I_2])

        m_o = m_I_1 * m_I_2
        # print(m_o)
        theta_bar = theta[O][1]

        numi += m_o * theta_bar
        denumi += m_o
    
    return numi / denumi

defuzz(18,200)
```

The following then calculates the next datapoint using x, \phi, and \theta.

```
# X and Phi extraction 
import math

def X(x,phi,theta):
    return x + math.cos(math.radians(phi)+math.radians(theta)) + math.sin(math.radians(phi)) * math.sin(math.radians(theta))

def PHI(phi,theta):
    return phi - math.degrees(math.asin((2 * math.sin(theta))/4))

```

At last, we can record each of the points visited for reaching the final position (10,90).

```

def track(x_test, phi_test):
    track_ls = []

    x_curr = x_test
    phi_curr = phi_test

    while True:
        theta = defuzz(x_curr, phi_curr)
        track_ls.append((x_curr,phi_curr, theta))
        # print(x_curr, phi_curr, theta)
        if math.fabs(x_curr - 10) < 2 and  math.fabs(phi_curr - 90) < 8:
            break
        x_curr = X(x_curr,phi_curr,theta)
        phi_curr = PHI(phi_curr, theta)
    
    return track_ls
```
